# Variablea aleatorias

Una variable aleatoria es un concepto fundamental en estadística y teoría de la probabilidad. Se define como una función que asigna un número real a cada resultado de un espacio muestral asociado con un experimento aleatorio. Las variables aleatorias se utilizan para cuantificar los resultados de procesos aleatorios y permiten trabajar con ellos desde un punto de vista matemático. Hay dos tipos principales de variables aleatorias: *discretas* y *continuas*.

### Funcion:

Una función en matemáticas es una relación entre dos conjuntos donde cada elemento del primer conjunto (dominio) se asigna a un único elemento del segundo conjunto (codominio). Se representa comúnmente como ( f(x) ), donde ( x ) es la variable independiente y ( f(x) ) es la variable dependiente.

## Variable Aleatoria Discreta

Una variable aleatoria *discreta* es aquella que puede tomar un número finito o contable de valores distintos. Estos valores suelen ser el resultado de contar y pueden representarse en un conjunto listable.

**Por ejemplo**, el número de caras obtenidas al lanzar dos monedas es una variable aleatoria discreta, ya que los posibles resultados (0, 1, 2 caras) son finitos y contables.

**Por ejemplo**, un contador presenta el examen para certificarse como contador público. El examen tiene cuatro partes. Defina una variable aleatoria $x$ como $x=$ número de partes del examen aprobadas.
$$x= 0, 1, 2, 3 , 4$$

**Por ejemplo**, considere el experimento de observar
los automóviles que llegan a una caseta de peaje. La variable aleatoria que interesa es $x =$ número de automóviles que llega a la caseta de peaje en un día. Los valores que puede tomar la
variable aleatoria son los de la secuencia $x = 0, 1, 2,$ etc. Así, $x$ es una variable aleatoria discreta que toma uno de los valores de esta sucesión infinita.

### ¿Es posible que una variable cualitativa se pueda interpretar como una variable aleatoria?

**Si**, pero es necesario hacer una asignación numérica.

**OJO**:  Los valores numéricos de esta variable son arbitrarios, pero son aceptables de acuerdo con la definición de una variable aleatoria.

**Ejemplo:**
En una encuesta se le puede preguntar a una persona si recuerda el mensaje de un comercial de televisión.

Este experimento tiene dos resultados: que la persona no recuerda el mensaje y que la persona recuerda el mensaje.

Estos resultados se describen numéricamente definiendo una variable aleatoria $x$ como sigue: sea $x=0$ si la persona no recuerda el mensaje y sea $x= 1$ si la persona recuerda el mensaje.

(podría haber usado 5 y 10)


### Variable Aleatoria Continua

Una variable aleatoria *continua*, por otro lado, puede tomar un número infinito de posibles valores, los cuales pueden representar rangos de valores dentro de un intervalo. Estos valores son el resultado de medir. Por ejemplo, el *tiempo* que tarda una persona en llegar al trabajo es una variable aleatoria continua, ya que teóricamente puede tomar cualquier valor dentro de un rango.



**Ejemplo,**

Considere el experimento de observar las llamadas telefónicas que llegan a la oficina de atención de una importante empresa de seguros. La variable aleatoria que interesa es $x=$ tiempo en minutos entre dos llamadas consecutivas.
¿Qué valores puede tomar $x$?

$$x\, \geq\, \, 0$$
$$x\, \in\, \, [0,\ \infty)$$


**Ejemplo,**

Considere el tramo de 90 millas de una carretera entre Atlanta y Georgia.
Para el servicio de ambulancia de emergencia en Atlanta, la variable aleatoria x es el  número de millas hasta el punto en que se localiza el siguiente accidente de tráfico en este tramo de la carretera.
¿Qué valores puede tomar $x$?

En este caso, $x$ es una variable aleatoria continua que toma cualquier valor en el intervalo $$0\, \, \leq\, x\, \leq\, 90.$$


# DISTRIBUCIONES DE PROBABILIDAD

Las variables aleatorias se describen matemáticamente mediante sus distribuciones de probabilidad, que asignan probabilidades a los diferentes posibles valores que la variable aleatoria puede tomar. Para variables aleatorias **discretas**, esta distribución se describe a menudo mediante una **función de masa de probabilidad (FMP)**, mientras que para variables aleatorias **continuas**, se utiliza una función de **densidad de probabilidad (FDP).**

Las distribuciones de probabilidad proporcionan herramientas importantes para calcular la probabilidad de que una variable aleatoria caiga dentro de un cierto rango de valores, así como para calcular medidas descriptivas como la media, la varianza y la desviación estándar, que ayudan a entender la naturaleza y el comportamiento de la variable aleatoria.

## Distribución Binomial

### Un experimento binomial

Un experimento binomial tiene las cuatro propiedades siguientes.

1. El experimento consiste en una serie de $n$ ensayos idénticos.
2. En cada ensayo hay dos resultados posibles. A uno de estos resultados se le llama *éxito* y al otro se le llama *fracaso*.
3. La probabilidad de éxito, que se denota $p$, no cambia de un ensayo a otro. Por ende, la probabilidad de fracaso, que se denota $1 - p$, tampoco cambia de un ensayo a otro.
4. Los ensayos son independientes.

La **función de masa de probabilidad de una distribución binomial** da la probabilidad de obtener exactamente "$x$" éxitos en "$n$" ensayos. La fórmula de la distribución binomial es:

$$P(X = x) = (nCx) \times p^x \times (1-p)^{n-x}$$


**Valor esperado**

$$E\left(x\right)=\mu=np $$
**Varianza**
$$\sigma^2=np(1-p) $$

# La distribución binomial negativa

La distribución binomial negativa es una distribución de probabilidad discreta que modela el número de ensayos necesarios para lograr un número especificado de éxitos en una serie de ensayos de Binomiales.

A diferencia de la distribución binomial, que cuenta el número de éxitos en un número fijo de ensayos, la distribución binomial negativa se centra en el número de ensayos necesarios para alcanzar un número fijo de éxitos.

### Parámetros:

La distribución binomial negativa tiene dos parámetros principales:

* $r$: El número de éxitos que se desea alcanzar.

* $p$: La probabilidad de éxito en cada ensayo.

$$P(X = n) = {n-1 \choose r-1} \times p^r \times (1-p)^{(n-r)}$$

Donde:

1. $P(X = n)$ es la probabilidad de que se necesiten exactamente "n" ensayos para obtener "$r$" éxitos.
2. ${n-1 \choose r-1}$ es el coeficiente binomial que representa el número de formas en que se pueden obtener "$r-1$" éxitos en "$n-1$" ensayos.
3. $p^r$ es la probabilidad de obtener "$r$" éxitos.
4. $(1-p)^{n-r}$ es la probabilidad de obtener "$n-r$" fracasos.

#### Usos comunes:

* Modelar tiempos de espera: Por ejemplo, cuántas veces debes lanzar un dado antes de obtener tu tercer seis.

* Ciencia de datos y estadística: En experimentos que involucran tasas de éxito/fallo, especialmente cuando el interés se centra en el número de intentos antes de alcanzar una meta.


#### Diferencias con la distribución binomial:

Mientras que la distribución binomial fija el número de ensayos y cuenta el número de éxitos, la distribución binomial negativa fija el número de éxitos y cuenta el número de ensayos necesario para alcanzar esos éxitos. Esto hace que la binomial negativa sea especialmente útil en situaciones donde el punto de interés es el esfuerzo o el tiempo requerido para alcanzar un objetivo, en lugar de la tasa de éxito dentro de un conjunto fijo de intentos.

```python
from scipy.stats import nbinom
```

**Parámetros de nbinom.pmf:**

*n*: Número fijo de éxitos. Es el número de éxitos que se buscan alcanzar.

*p*: Probabilidad de éxito en cada ensayo individual. Debe ser un número entre 0 y 1.

## Esperanza y varianza para la binomial negativa

La **esperanza** o media ($μ$) de una distribución binomial negativa se refiere al número promedio de ensayos necesarios para lograr un número fijo de éxitos ($r$), dada una probabilidad de éxito ($p$) en cada ensayo. Se calcula con la fórmula:
$$\mu = \cfrac{r}{p}$$

La **varianza** ($σ^2$) mide la dispersión de los valores de la variable alrededor de su media. En el caso de la distribución binomial negativa, la varianza se calcula como:

$$\sigma^2 =\cfrac{r(1-p)}{p^2}$$



### Distribución de Poisson

La **distribución de Poisson** es una distribución de probabilidad discreta que expresa la probabilidad de que ocurra un determinado número de eventos en un *intervalo fijo* de tiempo o espacio, bajo la premisa de que estos eventos ocurren con una tasa *media conocida* y de manera independiente de la hora desde el último evento. Es especialmente útil para modelar la cantidad de veces que ocurre un evento en un intervalo de tiempo determinado cuando los eventos ocurren de manera aleatoria y a una tasa constante.

### Características clave de la distribución de Poisson:

* **Eventos independientes:** Los eventos ocurren de manera independiente.
 La ocurrencia de un evento no afecta la probabilidad de otro.
* **Tasa constante:** Los eventos ocurren a una tasa media constante (λ) conocida.
* **Intervalos de tiempo o espacio fijos:** La distribución modela eventos dentro de intervalos de tiempo o espacio específicos y fijos.
* **Número de ocurrencias:** La distribución da la probabilidad de observar un número exacto de eventos dentro del intervalo dado.

### Función de masa de probabilidad (FMP):
La FMP de la distribución de Poisson para un número $k$ de eventos es dada por la fórmula:

$$P(X = k) = \frac{e^{-λ} \cdot λ^k}{ k!}$$

donde:

* $X$ es la variable aleatoria que representa el número de eventos.
* $k$ es el número de ocurrencias del evento (puede tomar valores 0, 1,2, ...).
* $e$ es la base del logaritmo natural (≈2.71828).
* $k!$ es el factorial de $k$.

En esta distribución

$$E(X) =\lambda = s^2 $$

### Ejemplos de uso de la distribución de Poisson:

* Llamadas telefónicas: Calcular la probabilidad de recibir un cierto número de llamadas en una central telefónica en una hora.
* Visitas a una página web: Estimar cuántas visitas se esperan en una página web en un día dado.
* Descomposición radiactiva: Modelar el número de desintegraciones radiactivas que ocurren en una muestra en un período de tiempo.
* Sistemas de colas: Analizar la llegada de clientes a un banco y su servicio.

# Distribución Hipergeométrica

La distribución hipergeométrica es una distribución de probabilidad discreta que describe la probabilidad de obtener exactamente $k$ éxitos en $n$ extracciones de una población finita de tamaño $N$ que contiene exactamente $K$ objetos exitosos, **sin reemplazo**.

A diferencia de la distribución binomial, donde cada ensayo es independiente y se realiza con reemplazo, la distribución hipergeométrica se aplica a situaciones donde los ensayos no son independientes, debido a que la extracción se realiza sin reemplazo.

image.png



**Parámetros:**

* $N$: El tamaño total de la población.
* $K$: El número total de objetos exitosos en la población.
* $n$: El número de extracciones realizadas.
* $k$: El número de éxitos observados (objetos exitosos extraídos).

**Características clave:**

* *Sin reemplazo*: A diferencia de la distribución binomial, los ensayos (extracciones) se realizan sin reemplazo, lo que significa que la probabilidad de éxito cambia después de cada extracción.
* *Dependencia entre ensayos*: La probabilidad de éxito en un ensayo depende de los resultados de los ensayos anteriores debido al mecanismo de extracción sin reemplazo.
* *Población finita*: La distribución se aplica a situaciones con una población finita y bien definida.

**Función de masa de probabilidad:**

La función de probabilidad de una variable aleatoria discreta con distribución hipergeométrica se define como:

$$ P(X = k) = \frac{{\binom{K}{k} \cdot \binom{N-K}{n-k}}}{{\binom{N}{n}}} $$

**Esperanza:**
$$E(X) = \frac{nK}{N}.$$

**Varianza:**
$$Var(X) = \frac{nK(N-K)(N-n)}{N^2(N-1)}$$

**Ejemplos de uso:**

* Control de calidad: Determinar la probabilidad de encontrar un cierto número de piezas defectuosas en una muestra aleatoria extraída de un lote de producción.
* Ecología: Estimar la probabilidad de capturar una cantidad específica de una especie particular en un número de capturas dadas, dentro de una población finita.
* Juegos de cartas: Calcular la probabilidad de obtener un número específico de ciertos tipos de cartas al repartir una mano de cartas de una baraja completa.

**Ejemplo**

Una empresa fabrica fusibles que empaca en cajas de 12 unidades cada una. Asuma que un inspector selecciona al azar tres de los 12 fusibles de una caja para inspeccionarlos.

$x=$ número de elementos defectuosos sacados por el inspector
$$x=0,\ 1,\ 2,\ 3 $$

Si la caja contiene exactamente cinco fusibles defectuosos, ¿cuál es la probabilidad de que el inspector encuentre que uno de los tres fusibles está defectuoso?



$N=12$


$K = 5$


$n=5$


$k=0,1,2,3$, para la pregunta particular $k=1$

$$ P(X = k) = \frac{{\binom{K}{k} \cdot \binom{N-K}{n-k}}}{{\binom{N}{n}}} $$

```python
comb(5, 1)*comb(7, 2)/comb(12, 3)
```

```python
from scipy.stats import hypergeom

# Parámetros de la distribución
N = 12  # Tamaño total de la población
K = 5  # Número total de éxitos en la población
n = 3   # Número de extracciones

# Calcular la FMP para obtener exactamente k=2 éxitos (cartas rojas)
k = 1
probabilidad = hypergeom.pmf(k, N, K, n)

print(f"Probabilidad de obtener {k} fusible dañado al sacar {n} fusibles: {probabilidad}")
```

```python
hypergeom.cdf(range(0, 4), 12, 5, 3)
```

```python
# Parámetros de la distribución
N = 12  # Tamaño total de la población
K = 5  # Número total de éxitos en la población
n = 3   # Número de extracciones

# Generar 10 muestras aleatorias de la distribución
muestras = hypergeom.rvs(N, K, n, size=10)
muestras
```

```python
import matplotlib.pyplot as plt

frecuencias = [list(muestras).count(i) for i in range(n+1)]

# Crear el diagrama de barras
plt.bar(range(n+1), frecuencias, color='red', alpha=0.7)
plt.xlabel('Variable aleatoria: Número de ')
plt.ylabel('Frecuencia')
plt.title('10 simulaciones de la Hipergeométrica')
plt.xticks(range(n+1))
plt.show()
```

Simular el proceso de extraer 5 cartas de una baraja estándar de 52 cartas en 10 ocasiones diferentes y contar cuántas cartas rojas aparecen en cada extracción, sin reemplazar las cartas entre las extracciones.

```python
# Parámetros de la distribución
N = 52  # Tamaño total de la población
K = 26  # Número total de éxitos en la población
n = 5   # Número de extracciones

# Generar 10 muestras aleatorias de la distribución
muestras = hypergeom.rvs(N, K, n, size=10)
muestras
```

```python
frecuencias = [list(muestras).count(i) for i in range(n+1)]

# Crear el diagrama de barras
plt.bar(range(n+1), frecuencias, color='red', alpha=0.7)
plt.xlabel('Número de Cartas Rojas')
plt.ylabel('Frecuencia')
plt.title('Frecuencia de Cartas Rojas en 10 Manos de 5 Cartas')
plt.xticks(range(n+1))
plt.show()
```
